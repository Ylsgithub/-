## 1.数值误差的避免

### (a).
设 $\epsilon=\frac{1}{2}\epsilon_m$，这里将$fl(x)=x(1+\delta)$中的$\delta$取为最大值$\epsilon$

则$\bigoplus_{i=1}^N x_i$

$=(((x_1+x_2)(1+\epsilon)+x_3)(1+\epsilon)+\cdots+x_N)(1+\epsilon)$

$=(x_1+x_2)(1+\epsilon)^{n-1}+x_3(1+\epsilon)^{N-2}+\cdots+x_N(1+\epsilon)$

从而$\frac{\sum_{i=1}^N x_i}{N}$的舍入结果为$\frac{(x_1+x_2)(1+\epsilon)^{n-1}+x_3(1+\epsilon)^{N-2}+\cdots+x_N(1+\epsilon)}{N}(1+\epsilon)$

于是$\bar{x}$的舍入误差为$\frac{(x_1+x_2)(1+\epsilon)^{n-1}+x_3(1+\epsilon)^{N-2}+\cdots+x_N(1+\epsilon)}{N}(1+\epsilon)-\frac{\sum_{i=1}^N x_i}{N}$

$=\frac{(x_1+x_2)n\epsilon+x_3(n-1)\epsilon+\cdots+x_N\cdot2\epsilon}{N}$

$\le x_{max}\frac{2n\epsilon+(n-1)\epsilon+(n-2)\epsilon+\cdots+2\epsilon}{N}$

$=\frac{x_{max}}{2}(n+3-\frac{2}{n})\epsilon$

$\simeq \frac{x_{max}(n+3)}{2}\cdot \frac{\epsilon_M}{2}$

## 3.Hilbert 矩阵

### (a).
由$D=\int_{0}^{1}dx(\sum_{i=1}^{n}c_{i}x^{i-1}-f(x))^2$，为使$D$取最小值，我们要$\frac{\partial D}{\partial c_{j}}=0$

即$\int_0^1dx\cdot 2(\sum_{i=1}^{n}c_{i}x^{i-1}-f(x))x^{j-1}=0$，$\sum_{i=1}^n c_i \int_0^1 x^{i+j-2}dx=\int_0^1x^{j-1}f(x)dx$

即$\sum_{i=1}^n \frac{1}{i+j-1}c_i=\int_0^1x^{j-1}f(x)dx$

所以$(H_n)_{ij}=\frac{1}{i+j-1}$，$b_i=\int_0^1x^{i-1}f(x)dx$

### (b).
${}c^tAc=\sum_{1\leq i,j\leq n}\frac{c_ic_j}{i+j-1}=\sum_{1\leq i,j\leq n}c_ic_j\int_0^1t^{i+j-2}dt=\int_0^1\left(\sum_{i=1}^n c_it^{i-1}\right)^2dt\ge0$

等号成立时有$c_i=0$，$i=1,2,3,\cdots$

对于一个实正定矩阵，其特征根都为正实数，故其行列式必定大于零，从而希尔伯特矩阵非奇异。

### (c).

由斯特林公式，$n!\simeq \sqrt{2 \pi n}(\frac{n}{e})^n$，两边取对数有$\ln(n!) \simeq \frac{1}{2}\ln(2\pi)+(n+\frac{1}{2})\ln(n)-n$，
于是得以快速估计$det(H_n)$近似值如下：

```python
import numpy as np
import pandas as pd

# 计算ln(n!)
def log_frac(n): return np.log(2 * np.pi) / 2 + (n + 1 / 2) * np.log(n) - n

# 计算ln(c_n)
def log_c(n): return sum([log_frac(i) for i in range(1, n)])

# 计算det(H_n)
def det_H(n): return np.exp(4 * log_c(n) - log_c(2 * n))

h_n=[det_H(i) for i in range(2,11)]
pd.DataFrame(columns=(["n"]+[i for i in range(2,11)]),data=[["H(n)"]+h_n])
```

### (d).
算法代码如下：

```python
# 上三角矩阵的反代求解
def Solve1(u, b):
    n = u.shape[0]
    x = np.zeros(n, dtype=float)
    x[n - 1] = b[n - 1] / u[n - 1, n - 1]
    for i in range(n - 2, -1, -1):
        x[i] = b[i]
        for j in range(i + 1, n):
            x[i] -= u[i, j] * x[j]
        x[i] /= u[i, i]
    return x


# 下三角矩阵的反代求解
def Solve2(u, b):
    n = u.shape[0]
    x = np.zeros(n, dtype=float)
    x[0] = b[0] / u[0, 0]
    for i in range(1, n):
        x[i] = b[i]
        for j in range(0, i):
            x[i] -= u[i, j] * x[j]
        x[i] /= u[i, i]
    return x


# Gauss消元法，将a变成上三角矩阵，
def GEM(aa, bb): 
    
    # 使用copy避免改变了原有数据
    a = np.copy(aa)
    b = np.copy(bb)
    n = a.shape[0]
    
    for i in range(0, n):
        
        # 支点遴选
        m = np.abs(a[i, i])
        id = i
        for j in range(i + 1, n):
            if np.abs(a[j, i]) > m: # 找最大的|a_ji|
                m = np.abs(a[i, j])
                id = j 
        if i != id: # 将两行互换
            a[[i, id], :] = a[[id, i], :]
            b[[i, id]] = b[[id, i]]
        
        # 进行消元
        for j in range(i + 1, n): # a_jk
            for k in range(i + 1, n): 
                a[j, k] -= a[i, k] * a[j, i] / a[i, i]
            b[j] -= b[i] * a[j, i] / a[i, i]
            a[j, i] = 0
    return (a, b)


# Gauss消元法解方程
def Solve_GEM(a, b):
    u, b = GEM(a, b) # 进行高斯消元
    return Solve1(u, b) # 反解下三角矩阵


# Cholesky分解，得到的h为下三角矩阵，其转置为上三角矩阵
def Cholesky(a):
    
    n = a.shape[0]
    h = np.zeros([n, n])
    h[0, 0] = np.sqrt(a[0, 0]) # h_11
    
    for i in range(1, n):
        
        for j in range(0, i): # h_ij
            h[i, j] = a[i, j]
            for k in range(0, j):
                h[i, j] -= h[i, k] * h[j, k]
            h[i, j] /= h[j, j]
            
        h[i, i] = a[i, i] 
        for k in range(0, i): # h_ii
            h[i, i] -= h[i, k] ** 2
        h[i, i] = np.sqrt(h[i, i])
    return h


# Cholesky分解解方程
def Solve_Cholesky(a, b):
    h = Cholesky(a) # 对a进行Cholesky分解
    return Solve1(np.transpose(h), Solve2(h, b)) # 分别反解一个上三角矩阵和下三角矩阵


# Hilbert矩阵的生成
def Hilbert(n):
    h = np.zeros([n, n], dtype=float)
    for i in range(0, n):
        for j in range(0, n):
            h[i, j] = 1 / (i + j + 1)
    return h

# b_n的生成
def b(n): return np.array([1. for i in range(n)])
```

计算结果如下：

(1)使用Gauss消元法的结果：

```python
colu=["阶数","x1","x2","x3","x4","x5","x6","x7","x8","x9","x10"]
result1=[[i]+np.ndarray.tolist(Solve_GEM(Hilbert(i),b(i))) for i in range(2,11)]
pd.DataFrame(columns=colu,data=result1)
```

(2)使用Cholesky分解的结果如下：

```python
result2=[[i]+np.ndarray.tolist(Solve_Cholesky(Hilbert(i),b(i))) for i in range(2,11)]
pd.DataFrame(columns=colu,data=result2)
```
